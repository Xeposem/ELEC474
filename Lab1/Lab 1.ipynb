# 5.2 Solution

```python
new_img_horizontal = np.zeros(img.shape, np.uint8)
```

```python
new_img_vertical = np.zeros(img.shape, np.uint8)
```

```python
plt.imshow(cv2.cvtColor(new_img_horizontal, cv2.COLOR_BGR2RGB))
plt.title("Lena Color")
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        new_img_horizontal[i][img.shape[1]-j-1]=img[i][j]
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        new_img_vertical[img.shape[0]-i-1][j]=img[i][j]
```

```python
gray_horizontal = cv2.cvtColor(new_img_horizontal, cv2.COLOR_BGR2GRAY)
gray_vertical = cv2.cvtColor(new_img_vertical, cv2.COLOR_BGR2GRAY)
```

```python
fig = plt.figure(figsize=(10,10))

fig.add_subplot(2, 2, 1)
#
plt.imshow(cv2.cvtColor(new_img_horizontal, cv2.COLOR_BGR2RGB))
plt.title("Lena Color Horizontal Flip")

fig.add_subplot(2, 2, 2)
#
plt.imshow(cv2.cvtColor(new_img_vertical, cv2.COLOR_BGR2RGB))
plt.title("Lena Color Vertical Flip")

fig.add_subplot(2, 2, 3)
#
plt.imshow(gray_horizontal, cmap='gray')
plt.title("Lena Gray Horizontal Flip")

fig.add_subplot(2, 2, 4)
#
plt.imshow(gray_vertical, cmap='gray')
plt.title("Lena Gray Vertical Flip")
```

# 5.3 Solution

```python
img_neg = np.zeros(img.shape, np.uint8)
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        for k in range(3):
            img_neg[i][j][k]=abs(256-img[i][j][k])
```

```python
plt.imshow(cv2.cvtColor(img_neg, cv2.COLOR_BGR2RGB))
plt.title("Lena Color Negated")
```

```python
gray_neg = cv2.cvtColor(img_neg, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_neg, cmap = 'gray')
plt.title("Lena Gray Negated")
```

# 5.3 Solution

```python
img_blue = np.zeros(img.shape, np.uint8)
img_green = np.zeros(img.shape, np.uint8)
img_red = np.zeros(img.shape, np.uint8)
```

```python
imgs = [img_blue,img_green,img_red]
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        for k in range(3):
            imgs[0][i][j][0]=img[i][j][0]
            imgs[1][i][j][1]=img[i][j][1]
            imgs[2][i][j][2]=img[i][j][2]
```

```python
fig = plt.figure(figsize=(10,10))

fig.add_subplot(1, 3, 1)
#
plt.imshow(cv2.cvtColor(img_blue, cv2.COLOR_BGR2RGB))
plt.title("Lena Blue Img")

fig.add_subplot(1, 3, 2)
#
plt.imshow(cv2.cvtColor(img_green, cv2.COLOR_BGR2RGB))
plt.title("Lena Green Img")

fig.add_subplot(1, 3, 3)
#
plt.imshow(cv2.cvtColor(img_red, cv2.COLOR_BGR2RGB))
plt.title("Lena Red Img")
```

```python
reimg = np.zeros(img.shape, np.uint8)
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        for k in range(3):
            reimg[i][j][0]=imgs[0][i][j][0]
            reimg[i][j][1]=imgs[1][i][j][1]
            reimg[i][j][2]=imgs[2][i][j][2]
```

```python
plt.imshow(cv2.cvtColor(reimg, cv2.COLOR_BGR2RGB))
plt.title("Merged Lena Img")
```
