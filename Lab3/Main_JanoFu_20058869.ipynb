```python
# To test out the code:
#step 1 : build the code
#step 2: adjust higher and lower threashold on the canny edge detection
#step 3: adjust the edge detection threshold trackbar, only change by 1 each time
#        At the first time changing, there should be a new window popping up about 
#        about a plot of canny vs general edge difference
#step 4: change a threshold by 1 until get a full graph
import cv2
import numpy as np
import copy
import math
import matplotlib.pyplot as plt

%matplotlib qt
img = cv2.imread('spiral.jpg')

CONST_THRESHOLD = 18
CONST_SMOOTHING = (5,5)
CONST_MIN = 7
CONST_MAX = 51

ratio = 0

plt.ion()


#To find and confirm the img size
img_rows = img.shape[0]
img_cols = img.shape[1]
img_size = img_rows * img_cols

#use grey picture as the sample in the lab instructions
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

ComparisionX =[]
ComparisionY = []

def GeneralEdgeDetection():
    #Defining the Sobel Mask, horizontal and vertical
    Sobel_Mask_V = np.array((
                  [-1, -2, -1],
                  [0,  0,  0],
                  [1,  2,  1]), dtype = "int")
    Sobel_Mask_H = np.array((
                  [-1,  0,  1],
                  [-2,  0,  2],
                  [-1,  0,  1]), dtype = "int")

    #Smoothing out the picture first (change to grey if wanted grey)
    SmoothRes = cv2.GaussianBlur(grey, CONST_SMOOTHING, cv2.BORDER_DEFAULT)

    #Applying Sobel Kernal Convolution to the filter
    SobelRes_H = cv2.filter2D(SmoothRes, cv2.CV_16S, Sobel_Mask_H)
    SobelRes_V = cv2.filter2D(SmoothRes, cv2.CV_16S, Sobel_Mask_V)
    GradientImage = GradientMagnitude(SobelRes_H, SobelRes_V)
    OutputEdge = BinaryThresholding(GradientImage)
    return OutputEdge
    
    
def GradientMagnitude(h, v):
    Gradientimg = grey.copy()
    for i in range (img_rows):
        for j in range (img_cols):
            Gradientimg[i][j] = math.hypot(h[i][j],v[i][j])
    return Gradientimg


def BinaryThresholding(GradientImg):
    Marked = np.zeros([img_rows,img_cols,1],dtype=np.uint8)
    for i in range (img_rows):
        for j in range (img_cols):
            if GradientImg[i][j] > CONST_THRESHOLD:
                Marked[i][j] = 255
    return Marked

def update(x):
    # update CONST_THRESHOLD when trackbar is moved
    global CONST_THRESHOLD
    global resG
    CONST_THRESHOLD = x
    resG = GeneralEdgeDetection()
    cv2.imshow("Gerneral Edge Detection Result", resG)
    difference = CompareCandG (resC,resG) 
    
    
    

def updateC_min(x):
    global resC
    global CONST_MIN
    CONST_MIN = x
    if CONST_MIN < CONST_MAX:
        SmoothRes = cv2.GaussianBlur(grey, CONST_SMOOTHING, cv2.BORDER_DEFAULT)
        resC = cv2.Canny(SmoothRes, CONST_MIN, CONST_MAX)
        print("\rPrinting Canny                                                         ", end = "")
        cv2.imshow("Canny Edge Detection Result", resC)
    else:
        print("\rmin is larger than max, min =",CONST_MIN, " max=", CONST_MAX, end = "")
        

def updateC_max(x):
    global resC
    global CONST_MAX
    CONST_MAX = x
    if CONST_MIN < CONST_MAX:
        SmoothRes = cv2.GaussianBlur(grey, CONST_SMOOTHING, cv2.BORDER_DEFAULT)
        resC = cv2.Canny(SmoothRes, CONST_MIN, CONST_MAX)
        print("\rPrinting Canny                                                          ", end = "")
        cv2.imshow("Canny Edge Detection Result", resC)
    else:
        print("\rmin is larger than max, min =",CONST_MIN, " max=", CONST_MAX, end= "")
        
def CompareCandG(c,g):
    
    global CONST_THRESHOLD
    global ratio
    similar = 0
    for i in range (img_rows):
        for j in range (img_cols):
            if (c[i][j] == g[i][j]):
                similar = similar+1
    ratio = similar / img_size
    ComparisionX.append(CONST_THRESHOLD)
    ComparisionY.append(ratio)
    plt.clf()
    plt.plot(ComparisionX,ComparisionY)
    plt.draw()
    plt.pause(0.001)
    plt.ioff()
            
    
```

```python
resG = GeneralEdgeDetection()
cv2.imshow("Gerneral Edge Detection Result", resG)
cv2.createTrackbar('Threshold = ','Gerneral Edge Detection Result',CONST_THRESHOLD,30,update)

resC = resG.copy()
updateC_max(CONST_MAX)
cv2.imshow("Canny Edge Detection Result", resC)
cv2.createTrackbar('Min = ','Canny Edge Detection Result',CONST_MIN,255,updateC_min)
cv2.createTrackbar('Max = ','Canny Edge Detection Result',CONST_MAX,255,updateC_max)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
