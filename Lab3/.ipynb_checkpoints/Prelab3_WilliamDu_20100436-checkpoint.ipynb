```python
def general_edge_detection(T):
    sobel_mask_x = np.array(([-1,0,1],[-2,0,2],[-1,0,1]), dtype = "int")
    sobel_mask_y = np.array(([-1,-2,-1],[0,0,0],[1,2,1]), dtype = "int")
    
    sobelx = cv.filter2D(smooth, cv.CV_16S, sobel_mask_x)
    sobely = cv.filter2D(smooth, cv.CV_16S, sobel_mask_y)
    
    res = np.zeros((img.shape[0], img.shape[1]), dtype = "int")
    
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            res[i][j] = math.sqrt((sobelx[i][j])**2 + (sobely[i][j])**2)
            if(res[i][j] > T):
                res[i][j] = 255
            else:
                res[i][j] = 0
    return res
```
